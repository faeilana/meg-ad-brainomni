```python
import os

data_dir = '/Users/faenegoro/Documents/Neuroimaging/meg_ad_project/data/meg_mri'

files = []
for f in os.listdir(data_dir):
    if f.endswith('_meg_rest_60sec.mat'):
        files.append(f)

labels = {}
for f in files:
    subject_id = f.split('_')[0]          # e.g. 'RSID0001'
    number = int(subject_id[4:])          # e.g. 1
    label = 0 if number <= 70 else 1      # 0 = control, 1 = AD
    labels[subject_id] = label

print(f"Total subjects: {len(labels)}")
print(f"Controls: {sum(v == 0 for v in labels.values())}")
print(f"AD: {sum(v == 1 for v in labels.values())}")


grad = data['grad']

grad_labels = [str(grad[0,0]['label'][i, 0][0]) for i in range(grad[0,0]['label'].shape[0])]
meg_labels = [str(data['label'][i, 0][0]) for i in range(data['label'].shape[0])]

print('grad channels:', len(grad_labels))
print('meg channels:', len(meg_labels))
print('first 5 grad labels:', grad_labels[:5])
print('first 5 meg labels:', meg_labels[:5])

# Find which grad indices match our 271 meg channels
matching_indices = [grad_labels.index(lbl) for lbl in meg_labels]
print('matching indices found:', len(matching_indices))
```
